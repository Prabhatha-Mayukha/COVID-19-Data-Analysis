```python
#number of cases in India

india_cases= covid[covid["location"]=="India"]
india_cases.head(20)
```

```python
#total cases per day
sns.set(rc={'figure.figsize':(20,10)})
sns.lineplot(x='date',y='total_cases',data=india_cases)
plt.show()
```

```python
india_cases_last_10_days= india_cases.tail(10)
```

```python
#last 10 days data
sns.set(rc={'figure.figsize':(20,10)})
sns.lineplot(x='date',y='total_cases',data=india_cases_last_10_days)
plt.show()
```

```python
#Total tests per day
sns.set(rc={'figure.figsize':(25,10)})
sns.lineplot(x="date",y="total_tests",data=india_cases)
plt.show()
```

```python
#Total tests in last 10 days
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="date",y="total_tests",data=india_cases_last_10_days)
plt.show()
```

```python
#Understanding cases of India, China and Japan
india_japan_china=covid[(covid["location"] =="India") | (covid["location"] =="China") | (covid["location"]=="Japan")]
```

```python
#Plotting growth of cases across China, India and Japan
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x="location",y="total_cases",data=india_japan_china)
plt.show()
```

```python
#Understanding cases of India, China and Japan
india_germany_spain=covid[(covid["location"] =="India") | (covid["location"] =="Germany") | (covid["location"]=="Spain")]
```

```python
#Plotting growth of cases across China, India and Japan
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x="location",y="total_cases",data=india_germany_spain)
plt.show()
```

```python
#Getting latest data
last_day_cases=covid[covid["date"]=="2020-05-24"]
last_day_cases
```

```python
#Sorting data w.r.t total_cases
max_cases_country=last_day_cases.sort_values(by="total_cases",ascending=False)
max_cases_country
```

```python
#Making bar-plot for countries with top cases
sns.barplot(x="location",y="total_cases",data=max_cases_country[1:6],hue="location")
plt.show()
```
